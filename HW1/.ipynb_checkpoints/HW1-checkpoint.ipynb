```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def hist_eq(img):
    L = 255
    # Initialize the histogram list
    histogram = []
    for i in range(L):
        histogram.append(0)

    # Get the shape of the image
    row, col = img.shape
    
    # Get the historgram
    for i in range(row):
        for j in range(col):
            count = histogram[img[i, j]]
            histogram[img[i, j]] = count + 1
    
    
    # Initialize a list to store transformed values     
    transformed_vals = []
    for i in range(L):
        transformed_vals.append(0)
    
    # Compute the transformed values     
    for i in range(L):
        sum = 0
        # Can optimize it here
        for j in range(i):
            sum = sum + histogram[j]    
        transformed_vals[i] = round(((L-1)/(row*col)) * sum)
        
    # Change the pixel value to its new value     
    for i in range(row):
        for j in range(col):
            img[i, j] = transformed_vals[img[i, j]]
    
    return img
```

```python
# ------------------------------------------------------
# Display the original image
print("Original Image")
img = cv2.imread('messi5.jpg', 0)
plt.imshow(img, 'gray')
plt.show()
cv2.imwrite("messi5_gray.jpg", img)

# ------------------------------------------------------
# Display the resulted image generated by the custom historgram equalization method
print("Custom Histogram Equalization")
img_copy = img.copy()
new_img = hist_eq(img_copy)
plt.imshow(new_img, 'gray')
plt.show()
cv2.imwrite("custom_histogram_eq_result.jpg", new_img)


# ------------------------------------------------------
# Display the resulted image generated by OpenCV method
print("OpenCV Histogram Equalization")
img_copy2 = img.copy()
new_img2 = cv2.equalizeHist(img_copy2)
plt.imshow(new_img2, 'gray')
plt.show()
cv2.imwrite("opencv_histogram_eq_result.jpg", new_img)
```

```python
import math

def convolution(img, kernel, kernel_sum):
    # Add padding
    kernel_height, kernel_width = kernel.shape
    
    # Get the pad size    
    pad_height = kernel_height // 2
    pad_width = kernel_width // 2
    
    # Pad the image
    img_pad = np.pad(img, ((pad_height, pad_height), (pad_width, pad_width)), 'constant', constant_values=0).astype(np.float32)
    
    new_img = np.zeros(img.shape)
    
    h = pad_height
    w = pad_width
    flatten_kernel = kernel.flatten()
    for i in range(h, img_pad.shape[0]-h):
        for j in range(w, img_pad.shape[1]-w):
            cur_section = img_pad[i-h:i-h+kernel_height, j-w:j-w+kernel_width]
            new_val = (cur_section.flatten() * flatten_kernel).sum()*kernel_sum
            new_img[i-h, j-h] = new_val

    return new_img.astype(np.uint8)

def gaussian_smoothing(img, filter_size, sigma):
    # Get the center index of the kernel     
    center_x = filter_size[0]//2
    center_y = filter_size[1]//2
    
    # Initialize the kernel     
    kernel = np.zeros((filter_size[0], filter_size[1]), np.float32)
    
    # To get the K value later
    sum = 0
    
    # 1st loop goes by rows     
    for i in range(filter_size[0]):
        # 2nd loop goes by cols
        for j in range(filter_size[1]):
            # Compute the weight value for each entry of the kernel            
            s = j - center_x
            r = i - center_y
            computed_val = math.exp(-((math.pow(s, 2)+math.pow(r, 2))/(2*math.pow(sigma, 2))))
            kernel[i, j] = computed_val
            sum = sum + computed_val
    
    # Compute the K value     
    K = 1/sum
    
    # RGB Image
    if len(img.shape) == 3:
        new_img = np.zeros(img.shape)
        for i in range(3):
            new_img[:,:,i] = convolution(img[:,:,i], kernel, K)
        return new_img.astype(np.uint8)
    
    # Gray Image
    else:
        return convolution(img, kernel, K).astype(np.uint8)
```

```python
# Display the original image
print("RGB")
print("Original Image")
img = cv2.imread('messi5.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()

# ------------------------------------------------------
# Custom Gaussian Blur
print("Custom Gaussian Blur")
img_copy3 = img_rgb.copy()
filter_size = (3,3)
sigma = 1
new_img3 = gaussian_smoothing(img_copy3, filter_size, sigma)
plt.imshow(new_img3)
plt.show()
new_img3_bgr = cv2.cvtColor(new_img3, cv2.COLOR_RGB2BGR)
cv2.imwrite("custom_gaussian_blur_bgr.jpg", new_img3_bgr)

# ------------------------------------------------------
# OpenCV Gaussian Blur
print("OpenCV Gaussian Blur")
img_copy4 = img_rgb.copy()
new_img4 = cv2.GaussianBlur(img_copy4, (3, 3), 0)
plt.imshow(new_img4)
plt.show()
new_img4_bgr = cv2.cvtColor(new_img4, cv2.COLOR_RGB2BGR)
cv2.imwrite("opencv_gaussian_blur_bgr.jpg", new_img4_bgr)
```

```python
# ------------------------------------------------------
# Display the original image
print("Gray")
print("Original Image")
img = cv2.imread('messi5.jpg', 0)
plt.imshow(img, 'gray')
plt.show()

# ------------------------------------------------------
# Custom Gaussian Blur
print("Custom Gaussian Blur")
img_copy5 = img.copy()
filter_size = (3,3)
sigma = 1
new_img5 = gaussian_smoothing(img_copy5, filter_size, sigma)
plt.imshow(new_img5, 'gray')
plt.show()
cv2.imwrite("custom_gaussian_blur_gray.jpg", new_img5)

# ------------------------------------------------------
# OpenCV Gaussian Blur
print("OpenCV Gaussian Blur")
img_copy6 = img.copy()
new_img6 = cv2.GaussianBlur(img_copy6, (3, 3), 0)
plt.imshow(new_img6, 'gray')
plt.show()
cv2.imwrite("opencv_gaussian_blur_gray.jpg", new_img6)
```
