```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loaded the images
white_img = cv2.imread("white_img.jpg")
black_img = cv2.imread("black_img.jpg")
mask_img = cv2.imread("mask_img.jpg")
```

```python
# Builds a gaussian pyramid with the max_itrs layers on the given image
def buildGaussianPyramid(src_img, max_itrs):
    layer = src_img.copy().astype('float32')
    gaussian_pyramid = [layer]
    for i in range(max_itrs - 1):
        layer = cv2.pyrDown(layer)
        gaussian_pyramid.append(layer)
    return gaussian_pyramid

# Builds a laplacian pyramid using the passed-in gaussian pyramid
def buildLaplacianPyramid(g_pyr):
    lap_pyr = []
    lvls = len(g_pyr)
    
    for i in range(lvls - 1):
        size = (g_pyr[i].shape[1], g_pyr[i].shape[0])
        layer = g_pyr[i] - cv2.pyrUp(g_pyr[i+1], dstsize=size)
        lap_pyr.append(layer)
    
    # Last layer of laplacian = last layer of gaussian     
    lap_pyr.append(g_pyr[lvls-1])
    return lap_pyr
```

```python
# Set the number of layers
max_vals = 5
white_img_g_pyr = buildGaussianPyramid(white_img, max_vals)
print("nlevels of white_img_g_pyr: " + str(len(white_img_g_pyr)))

white_img_lap_pyr = buildLaplacianPyramid(white_img_g_pyr)
print("nlevels of white_img_lap_pyr: " + str(len(white_img_lap_pyr)))

# Display the laplacian images of the white image
for i in range(max_vals):
    layer = white_img_lap_pyr[i]
    plt.title('layer: '+str(i))
    print('lap_pyr.shape:', layer.shape)
    
    # Scaling to display images; values must be within 0-255     
    scale_factor = 255.0/np.max(layer)
    layer = cv2.convertScaleAbs(layer, alpha=scale_factor)
    layer = cv2.cvtColor(layer, cv2.COLOR_BGR2RGB)
    plt.imshow(layer)
    plt.show()
```

```python
black_img_g_pyr = buildGaussianPyramid(black_img, max_vals)
print("nlevels of black_img_g_pyr: " + str(len(black_img_g_pyr)))

black_img_lap_pyr = buildLaplacianPyramid(black_img_g_pyr)
print("nlevels of black_img_lap_pyr: " + str(len(black_img_lap_pyr)))

# Display the laplacian images of the black image
for i in range(max_vals):
    layer = black_img_lap_pyr[i]
    plt.title('layer: '+str(i))
    print('lap_pyr.shape:', layer.shape)
    
    scale_factor = 255.0/np.max(layer)
    layer = cv2.convertScaleAbs(layer, alpha=scale_factor)
    layer = cv2.cvtColor(layer, cv2.COLOR_BGR2RGB)
    plt.imshow(layer)
    plt.show()
```

```python
mask_img_g_pyr = buildGaussianPyramid(mask_img, max_vals)
print("nlevels of mask_img_g_pyr: " + str(len(mask_img_g_pyr)))

# Display the gaussian images of the mask image
for i in range(max_vals):
    layer = mask_img_g_pyr[i]
    plt.title('layer: '+str(i))
    print('g_pyr.shape:', layer.shape)
    
    scale_factor = 255.0/np.max(layer)
    layer = cv2.convertScaleAbs(layer, alpha=scale_factor)
    layer = cv2.cvtColor(layer, cv2.COLOR_BGR2RGB)
    plt.imshow(layer)
    plt.show()
```

```python
# Equation: blended_lap = mask_g * img1_lap + (1 - mask_g) * img2_lap
# (1 - mask_g) - > (255 - mask_g); changing 1 to 255 since the values range from 0 to 255 not 0 to 1
def compute_composite_image(white_lap, black_lap, mask_g):
    return (mask_g * white_lap) + ((255 - mask_g) * black_lap)
#     return cv2.add(cv2.multiply(mask_img, img1), cv2.multiply((1 - mask_img), img2))
    

def blend(white_lap_Pyr, black_lap_Pyr, maks_g_Pyr):
    blended_Pyr = []
    
    for i in range(len(maks_g_Pyr)):
        blended_Pyr.append(compute_composite_image(white_lap_Pyr[i], black_lap_Pyr[i], maks_g_Pyr[i]))
        
    return blended_Pyr
```

```python
blended_pyr = blend(white_img_lap_pyr, black_img_lap_pyr, mask_img_g_pyr)

for i in range(max_vals):
    layer = blended_pyr[i]
    plt.title('layer: '+str(i))
    print('lap_pyr.shape:', layer.shape)
    
    scale_factor = 255.0/np.max(layer)
    layer = cv2.convertScaleAbs(layer, alpha=scale_factor)
    layer = cv2.cvtColor(layer, cv2.COLOR_BGR2RGB)
    plt.imshow(layer)
    plt.show()
```

```python
def collapse(blended_Pyr):
    pyr_size = len(blended_Pyr)
    result = blended_Pyr[pyr_size-1]
    for i in range(pyr_size-1, 0, -1):
        size = (blended_Pyr[i-1].shape[1], blended_Pyr[i-1].shape[0])
        result = cv2.pyrUp(result, dstsize=size) + blended_Pyr[i-1]
    return result
```

```python
result_img = collapse(blended_pyr)
plt.title('Result Image')

scale_factor = 255.0/np.max(result_img)
result_img = cv2.convertScaleAbs(result_img, alpha=scale_factor)
result_img = cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB)
plt.imshow(result_img)
plt.show()

# print(g_pyr[len(g_pyr)-1])
# print(1 - g_pyr[len(g_pyr)-1])
temp = 1 - g_pyr[len(g_pyr)-1]
print(temp.shape)
# plt.imshow(1 - g_pyr[len(g_pyr)-1])
# plt.show()
```
