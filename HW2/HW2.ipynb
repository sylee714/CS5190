```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('fig1.jpg', 0)
plt.figure(figsize=(10,10))
plt.imshow(img, 'gray')
plt.show()

ret, img2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
plt.figure(figsize=(10,10))
plt.imshow(img2, 'gray')
plt.show()

ret, img3 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
plt.figure(figsize=(10,10))
plt.imshow(img3, 'gray')
plt.show()
```

```python
# Using Binary Thresholding
contours, hierarchy = cv2.findContours(img2.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

draw_img = np.zeros_like(img2)
print(draw_img)
res = cv2.drawContours(draw_img, contours, -1, (255, 255, 0), 2)
plt.figure(figsize=(10, 10))
plt.imshow(res, 'gray')
plt.show()
print(len(contours))
```

```python

low_thres = 50
high_thres = 200
canny_img = cv2.Canny(img2, low_thres, high_thres)
plt.figure(figsize=(10, 10))
plt.imshow(canny_img, 'gray')
plt.show()
contours2, hierarchy = cv2.findContours(canny_img.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

draw_img2 = np.zeros_like(img2)
res2 = cv2.drawContours(draw_img2, contours2, -1, (255, 255, 0), 2)
plt.figure(figsize=(10, 10))
plt.imshow(res2, 'gray')
plt.show()
```
